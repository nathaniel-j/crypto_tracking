```python
coin = pd.read_csv('crypto_market_cap_shares - Sheet1.csv')
coin.fillna(0, inplace=True)
coin
```

```python
plt.plot(coin['btc'], coin['year'], fillstyle='full')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(8,14))

y = coin['year']

eth_x = coin['eth']
ltc_x = eth_x + coin['ltc']
xrp_x = ltc_x + coin['xrp']
btc_x = xrp_x + coin['btc']
other_x = btc_x + coin['other']
bch_x = other_x + coin['bch']
eos_x = bch_x + coin['eos']

plt.plot(eos_x, y, label='EOS')
plt.fill_betweenx(y,eos_x, 0)

plt.plot(bch_x, y, label='BCH')
plt.fill_betweenx(y,bch_x, 0)

plt.plot(other_x, y, label='Other')
plt.fill_betweenx(y,other_x, 0)

plt.plot(btc_x, y, label='BTC')
plt.fill_betweenx(y,btc_x, 0)

plt.plot(xrp_x, y, label='XRP')
plt.fill_betweenx(y,xrp_x, 0)

plt.plot(ltc_x, y, label='LTC')
plt.fill_betweenx(y,ltc_x, 0)

plt.plot(eth_x, y, label='ETH')
plt.fill_betweenx(y,eth_x, 0)


plt.title('Cryptocurrencies by market cap share since 2013', fontsize=20, y=1.03, )
plt.legend(loc=9, fontsize=15, title='cryptocurrency')
plt.ylim(["Dec '21", "Jun '13"])
plt.xlim([0,100])
plt.show()
```
